```python
num_data_years = num_data[['YearBuilt', 'YearRemodAdd', 'YrSold']]
```

```python
fig, ax = plt.subplots(1, 3, figsize = (12, 8))
for idx, axis in enumerate(ax.flat):
    axis.scatter(num_data_years.iloc[:, idx], y)
    axis.set_xlabel(num_data_years.columns[idx])
plt.tight_layout()
```

```python
initial_val = 0.1
quantile_years = []
for i in range(1, 11):
    quantile_years.append(num_data_years['YearBuilt'].quantile(initial_val * i))
```

```python
quantile_years
```

```python
quantile_years
new_quantile_years=[]
for i in quantile_years:
    i = int(i)
    new_quantile_years.append(i)


new_quantile_years
```

```python
for idx, val in enumerate(new_quantile_years):
    vals = idx * 0.1
    cond = (num_data_years['YearBuilt'] <= val) & (num_data_years['YearBuilt'] >=1)
    num_data_years['YearBuilt'][cond] = vals
```
